```python
# group by industry code
df_grouped = df.groupby('industry_code')
# choose 7 as an example cause it has the most data
df_7 = df_grouped.get_group(7)
```

```python
# get each stock_id as columns and retrieve the 21d_next_raw as rows
df_7_pivot = df_7.pivot(columns='stock_id', values='21d_next_raw')
```

```python
# check the na in each columns in df_7_pivot
na_count = df_7_pivot.isna()
a = na_count.sum()
a.describe()
```

```python
# remove the columns with more than 80% na
df_7_pivot_d = df_7_pivot.dropna(axis=1, thresh=0.6*len(df_7_pivot))
df_7_pivot_d.isna().sum().describe()
```

```python
# check the na by index
na_count = df_7_pivot.isna()
a = na_count.sum(axis=1)
a.describe()
```
